# The Bar Plot

**alternative names**: bar graph / column chart <br>

**aesthetics**: lenght of bars, colours <br>

**variations**: vertical, horizontal, stacked, grouped<br>

**potential pitfalls**: not ideal for trends or intra-period fluctuations; axes that do not start from zero<br>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sitzungsgelder = pd.read_csv ('Sitzungsgelder2019-2021.csv')
sitzungsgelder
```

```python
width = 20
height = 10
sns.set(rc = {'figure.figsize':(width,height)})

barchart = sns.barplot(data=sitzungsgelder, 
            y="2021", 
            x="Fraktion",
            width=0.5,
            color='blue')

barchart.axes.set_title("Attendance fees of the Cologne City Council 2021 in € by political group",fontsize=30)
```

```python
# convert the dataframe to a long format
df_sitzungsgelder = pd.melt(sitzungsgelder, id_vars="Fraktion", var_name="year", value_name="money")
df_sitzungsgelder
```

```python
width = 20
height = 10
sns.set(rc = {'figure.figsize':(width,height)})

barchart_large = sns.barplot(data=df_sitzungsgelder, 
            y="money", 
            x="Fraktion",
            hue="year",
            width=0.5)

for i in barchart.containers:
    barchart.bar_label(i,)

barchart_large.axes.set_title("Attendance fees of the Cologne City Council from 2019-2021 in € by political group",fontsize=30)
```
